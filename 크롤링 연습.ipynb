### 크롤링 연습

#### yes24 베스트셀러(국내도서) 순위 크롤링

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'http://www.yes24.com/24/Category/BestSeller'
res = req.get(url)
res
```

```python
soup = bs(res.text,'lxml')
soup
```

```python
title = soup.select('p.image+p>a')
title
```

```python
len(title)
```

```python
author = soup.select('p.aupu>a:nth-child(1)')
author
```

```python
len(author)
```

```python
title_list = []
author_list = []
rank_list = []

for i in range(40):
    title_list.append(title[i].text)
    author_list.append(author[i].text)
    rank_list.append(i+1)

dic = {'순위':rank_list,'도서명':title_list,'작가':author_list}
df = pd.DataFrame(dic)
df.set_index('순위',inplace=True)
df
```
