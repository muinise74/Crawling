# Image Crawling

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup as bs
# 파일 시스템을 위한 라이브러리 : 파일 폴더 생성
import os
# 이미지의 경로를 파일로 저장시켜주는 라이브러리
from urllib.request import urlretrieve
```

```python
# 폴더를 생성 : 바탕화면에 이미지를 폴더가 없다면 폴더를 생성하기
if not os.path.isdir('C:/Users/smhrd/Desktop/이미지'):
    os.mkdir("C:/Users/smhrd/Desktop/이미지")
```

```python
driver = wb.Chrome()
driver.get("https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EA%B3%A0%EC%96%91%EC%9D%B4")
```

```python
soup = bs(driver.page_source,'lxml')
soup
```

```python
img = soup.select("img._image._listImage")
```

```python
imgSrc = []
for i in range(len(img)):
    imgSrc.append(img[i]['src'])
imgSrc
```

```python
for i in range(len(imgSrc)):
    urlretrieve(imgSrc[i],"C:/Users/smhrd/Desktop/이미지/Image"+str(i+1)+".jpg")
    time.sleep(1)
```

```python
driver.quit()
```
