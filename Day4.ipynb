```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

# YouTube에서 영상의 제목과 조회수 Crawling

```python
driver = wb.Chrome();
time.sleep(0.5)
driver.get("https://www.youtube.com/c/%EA%B8%B1%EB%B8%94Geekble/videos")
time.sleep(1)
# 스크롤 하는 방법
body = driver.find_element_by_css_selector("body")
for i in range(20):
    body.send_keys(Keys.END)
    time.sleep(2)
```

```python
title = driver.find_elements_by_id("video-title")
title
```

```python
title_list = []
for i in title:
    title_list.append(i.text)
title_list
```

```python
countClick = driver.find_elements_by_css_selector('div#metadata-line>span:nth-child(1)')
countClick
```

```python
countClick_list = []
for i in countClick:
    countClick_list.append(i.text)
countClick_list
```

```python
print(len(title),len(countClick))
```

```python
dic = {'Title':title_list,'조회수':countClick_list}
df_youtube = pd.DataFrame(dic)
df_youtube
```

```python
driver.quit()
```

# Gmarket Best 상품 수집

```python
driver = wb.Chrome();
time.sleep(0.5)
driver.get("http://corners.gmarket.co.kr/Bestsellers")
```

```python
name = []
price = []
```

```python
for i in range(100):
    img = driver.find_elements_by_css_selector('img.lazy')    
    img[i].click()
    time.sleep(2)    
    temp1 = driver.find_element_by_css_selector('h1.itemtit')
    temp2 = driver.find_element_by_css_selector('strong.price_real')
    name.append(temp1.text)
    price.append(temp2.text)
    driver.back()
    time.sleep(2)   
```

```python
print(name,price)
```

```python
dic = {'상품명': name, '가격':price}
dfGmarketBest = pd.DataFrame(dic)
dfGmarketBest
```

```python
dfGmarketBest.to_csv("GmarketBest.csv",encoding = 'euc-kr')
```

```python
driver.quit()
```
